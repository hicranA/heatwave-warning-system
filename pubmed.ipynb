```python
import requests

def search_pubmed(keyword):
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi"
    params = {
        "db": "pubmed",
        "term": keyword,
        "retmode": "json"
    }
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        pmids = data["esearchresult"]["idlist"]
        return pmids
    else:
        print(f"Error {response.status_code}: {response.text}")
        return []
```

```python
keyword = "heat"  # Replace with your desired keyword
articles = search_pubmed(keyword)
print(f"Found {len(articles)} articles related to {keyword}.")
print("PMIDs:", articles)
```

```python
import requests

def search_pubmed_for_country(country):
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi"
    queries = [
        f"\"heatwave warning system\" AND \"{country}\"",
        f"\"heat health warning system\" AND \"{country}\""
    ]
    
    for query in queries:
        params = {
            "db": "pubmed",
            "term": query,
            "retmode": "json"
        }
        
        response = requests.get(base_url, params=params)
        data = response.json()
        article_count = int(data["esearchresult"]["count"])
        
        if article_count > 0:
            return True  # Found articles related to the query for the country
    
    return False  # No articles found for any of the queries for the country

countries = ["India", "USA", "Canada", "Australia"]
results = {}

for country in countries:
    has_system = search_pubmed_for_country(country)
    results[country] = "Has system" if has_system else "No system found"

print(results)
```
