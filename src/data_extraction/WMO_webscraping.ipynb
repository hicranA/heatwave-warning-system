### Scraping from https://pubmed.ncbi.nlm.nih.gov/

```python
# Set up Chrome WebDriver
option = webdriver.ChromeOptions()
driver = webdriver.Chrome(options = option)

# Navigate to PubMed website
driver.get('https://pubmed.ncbi.nlm.nih.gov/')
```

```python
# Enter the search query in the search bar
search_bar = driver.find_element(By.CSS_SELECTOR, "input[type='search']")
search_bar.send_keys("heat wave")

# Click the search button
search_button = driver.find_element(By.CSS_SELECTOR, "span[class='usa-search-submit-text']")
search_button.click()

# Wait for the search results to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "a.docsum-title")))

# Extract the titles and links of the search results
titles = []
links = []
for article in driver.find_elements(By.CSS_SELECTOR, "a.docsum-title"):
    titles.append(article.text)
    links.append(article.get_attribute("href"))

# Save the scraped data to a CSV file
with open("pubmed_results.csv", "w", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["title", "link"])
    for i in range(len(titles)):
        writer.writerow([titles[i], links[i]])

# Close the WebDriver instance
driver.quit()

# Print the CSV file to the console
with open("pubmed_results.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
